```python
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
```

```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\n'

data = data.split(split_tag)
print(len(data))
data[-1]
```

```python
for d in data:
    if len(d) >0:
        name, image = d.split('\t')
        print(name, image)
```

## 將 txt 轉成 pandas dataframe

```python
import pandas as pd

arrange_data = {'name': [],
                'image': []}
for d in data:
    if len(d) > 0:
        name, image = d.split('\t')
        arrange_data['name'].append(name)
        arrange_data['image'].append(image)
df = pd.DataFrame(arrange_data)
df.head()
```
